## Distribución normal 

### Distribución normal estandar



### Ejemplo 1

```python
mu = 0
sigma = 1

Z = stats.norm(loc = mu, scale = sigma)
```

```python
z = np.arange(-4,4.05,0.01)

# FDP
f = Z.pdf(z)

fig, ax1 = plt.subplots()
ax1.plot(z,f)
ax1.set(xlabel = "z",
       ylabel ="f(z)",
       title = "Funcion de probabilidad estandar"
       )
ax1.grid()
```

```python
# FDA
F = Z.cdf(z)

fig, ax2 = plt.subplots()
ax2.plot(z,F)
ax2.set(xlabel = "z",
       ylabel ="F(z)",
       title = "Funcion de probabilidad acumulativa"
       )
ax2.grid()
```

Encuentre la proporcion de las observaciones de la distribución Normal estándar que son menores que -1.25:

$$P(Z \leq -1.25)$$

```python
# P1 = P(z <= -1.25)
z1 = -1.25
P1 = Z.cdf(z1)  # P1 = F(z1)
print(f"P(Z <= -1.25) = {P1:.4f}")
```

```python
# Grafica
fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(0, max(f) + 0.01)
# Limites
x_i = plt.vlines(x = -1.25, ymin = 0,ymax = Z.pdf(-1.25), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana normal')
plt.ylabel('$f(z)$')
plt.xlabel('z')
plt.plot(z, f,color='r')
ax.fill_between(z, f,0, where=(z<=-1.25), color='blue', alpha=.1)

plt.show()
```

Supongamos que queremos encontrar la proporción de observaciones de la distribución Normal estándar que son superiores a 0.81.

$$P(Z \geq 0.81) = 1 - P(Z < 0.81) = 1- P(Z \leq 0.81) = 1 - F(0.81)$$

```python
# P2 = P(z => 0.81)
z2 = 0.81
P2 = 1 - Z.cdf(z2) # P2 = 1 - F(z2)
print(f"P(z => 0.81) = {P2:.4f}")
```

```python
# Grafica
fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(0, max(f) + 0.01)
# Limites
x_i = plt.vlines(x = 0.81, ymin = 0,ymax = Z.pdf(0.81), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana normal')
plt.ylabel('$f(z)$')
plt.xlabel('z')
plt.plot(z, f,color='r')
ax.fill_between(z, f,0, where=(z>=0.81), color='blue', alpha=.1)

plt.show()
```

Encuentre la proporción de observaciones de la distribución Normal estándar que están entre -1.25 y 0.81.

$$P(-1.25 < Z < 0.81) = P(Z \leq 0.81) - P(Z \leq -1.25) = F(0.81) - F(-1.25)$$

```python
# Grafica
fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(0, max(f) + 0.01)
# Limites
x_i = plt.vlines(x = -1.25, ymin = 0,ymax = Z.pdf(-1.25), color = 'b', linestyles='--')
x_f = plt.vlines(x = 0.81, ymin = 0,ymax = Z.pdf(0.81), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana normal')
plt.ylabel('$f(z)$')
plt.xlabel('z')
plt.plot(z, f,color='r')
ax.fill_between(z, f,0, where=(z>=-1.25)&(z<=0.81), color='blue', alpha=.1)

plt.show()
```

```python
z1 = -1.25
z2 = 0.81
# P3 = P(−1.25 < z < 0.81)
P3 = Z.cdf(z2) - Z.cdf(z1) # P3 = F(z2) - F(z1)
print(f"P(−1.25 < z < 0.81) = {P3:.4f}")
```

## Solución de problemas que involucran distribuciones normales
Los siguientes pasos resumen el procedimiento.
1. **Plantee el problema** en términos de la variable observada $x$.
2. **Estandarizar y hacer la grafica estandarizada**. Estandarice $x$ para reformular el problema en términos de una variable Normal estándar $z$. Luego, haga un dibujo para mostrar el área de interés bajo la curva Normal estándar.
3. **Encuentre el área requerida** bajo la curva normal estándar usando la **tabla A** o cualquier software especializado
4. **Conclusión**. Escriba la conclusión en el contexto del problema.

### Ejemplo 1 - El tigre (Empleando normalización)

En el campo de prácticas, Tiger Woods practica su swing con un palo en particular golpeando muchas, muchas bolas. Cuando Tiger golpea a su driver, la distancia que recorre la pelota sigue una distribución Normal con una media de 304 yardas y una desviación estándar de 8 yardas. ¿Qué porcentaje de los golpes de Tiger recorren al menos 290 yardas?

**Paso 1 – Planteamiento del problema**

Sea $X$ la VA que define la distancia que la bola del Tiger viaja. Tenemos que $X \sim N(\mu,\sigma)$ donde:
* $\mu = 304$
* $\sigma = 8$

Luego $X \sim N(\mu = 304,\sigma = 8)$. Se pide: $P(X \geq 290)= ?$

**Paso 2 – Estandarizar y hacer la grafica estandarizada**

Hallemos la VA $Z$ tal que $Z \sim N(\mu = 0,\sigma = 1)$:

$$z=\frac{x-\mu}{\sigma} = \frac{x-304}{8}$$

De modo que para $x = 290$ tenemos que:

$$z = \frac{290-304}{8} = -1.75$$

De modo que lo que nos toca hayar es $P(z>=-1.75)$. 

```python
# Grafica
fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(0, max(f) + 0.01)
# Limites
x_i = plt.vlines(x = -1.75, ymin = 0,ymax = Z.pdf(-1.75), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana normal')
plt.ylabel('$f(z)$')
plt.xlabel('z')
plt.plot(z, f,color='r')
ax.fill_between(z, f,0, where=(z>=-1.75), color='blue', alpha=.1)

plt.show()
```

**Paso 3 – Encuentre el área requerida**

En resumen, lo que se ha hecho hasta el momento es:
* $x = 290 \to z = -1.75$
* $P(x \geq 290) = ? \to P(z \geq -1.75) = ?$

```python
# Parametros
mu = 304
sigma = 8

# Variable de interes
x1 = 290

# Normalizacion
z1 = (x1 - 304)/8 # z1 = -1.75

# P1 = P(z >= -1.75) = 1 - P(x <= -1.75) = 1 - F(-1.75)

P1 = 1 - Z.cdf(z1)
print(f"P(z >= {z1:.2f}) = {P1:.4f}")
```

**Paso 4 – Conclusión**

$$P(z \geq -1.75) = 0.9599 \to P(x \geq 290) = 0.9599$$

**Conclusión**: Alrededor del 96 % de las pelotas de Tiger Woods viajan al menos 290 yardas.

### Ejemplo 1 - El tigre (Sin normalizar)

**Planteamiento del problema**

Sea $X$ la VA que define la distancia que la bola del Tiger viaja. Tenemos que $X \sim N(\mu,\sigma)$ donde:
* $\mu = 304$
* $\sigma = 8$

Luego $X \sim N(\mu = 304,\sigma = 8)$. Se pide: $P(X \geq 290)= ?$

**Solucion usando python**

* **Definición de la VA**: $X \sim N(\mu = 304,\sigma = 8)$

```python
mu = 304
sigma = 8

# Variable Aleatoria: X ~ N(mu = 304, sigma = 8)
X = stats.norm(loc = mu, scale = sigma)
```

```python
x = np.arange(mu-3*sigma,mu+3*sigma,0.01)
f_x = X.pdf(x)
F_x = X.cdf(x)
```

```python
#plot 1: PDF

plt.subplot(2, 1, 1)
plt.title('Distribución Normal')
plt.ylabel('$f(x)$')
plt.xlabel('x')
plt.plot(x, f_x,color='r')

#plot 2: CDF
cdf_x = X.cdf(x)
plt.subplot(2, 1, 2)
plt.ylabel('$F(x)$')
plt.xlabel('x')
plt.plot(x, F_x,color='b')
plt.show()
```

* **Calculo de la probabilidad solicitada**: $P(x \geq 290)$

$$P(x \geq 290) = 1 - P(x \leq 290) = 1 - F(290)$$

```python
# P1 = P(x >= 290) = 1 - P(x <= 290) = 1 - F(290)
x1 = 290
P1 = 1 - X.cdf(x1)
print(f"P(x <= 290) = {P1:.4f}")
```

```python
# Grafica

fig, ax = plt.subplots()
ax.set_xlim(mu-3*sigma,mu+3*sigma)
ax.set_ylim(0, max(f_x) + 0.01)
# Limites
x_i = plt.vlines(x = 290, ymin = 0,ymax = X.pdf(290), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana')
plt.ylabel('$f(x)$')
plt.xlabel('x')
plt.plot(x, f_x,color='r')
ax.fill_between(x, f_x,0, where=(x>=290), color='blue', alpha=.1)

plt.show()
```

## Ejemplo 2 - El tigre (Sin normalizar)

¿Qué porcentaje de los pelotas golpeadas por Tiger recorren entre 305 y 325 yardas?

**Planteamiento del problema**

Sea $X$ la VA que define la distancia que la bola del Tiger viaja. Tenemos que $X \sim N(\mu,\sigma)$ donde:
* $\mu = 304$
* $\sigma = 8$

Luego $X \sim N(\mu = 304,\sigma = 8)$. Se pide: $P(305 \geq X \leq 325)= ?$

$$P(305 \geq X \leq 325) = P(x \leq 305) - P(x \leq 305) = F(305) - F(325)$$

```python
# P2 = P(305 < x < 0.325)
P2 = X.cdf(325) - X.cdf(305) # P3 = F(325) - F(305)
print(f"P(305 < x < 0.325) = {P2:.4f}")

fig, ax = plt.subplots()
ax.set_xlim(mu-3*sigma,mu+3*sigma)
ax.set_ylim(0, max(f_x) + 0.01)
x_i = plt.vlines(x = 305, ymin = 0,ymax = X.pdf(305), color = 'b', linestyles='--')
x_f = plt.vlines(x = 325, ymin = 0,ymax = X.pdf(325), color = 'b', linestyles='--')
plt.title('Distribución Gaussiana')
ax.plot(x, f_x)
ax.set(ylabel='f')
ax.fill_between(x, f_x,0, where=(x>=305)&(x<=325), color='blue', alpha=.1)
ax.grid()
```

**Conclusión**

$$P(305 \geq X \leq 325) = 0.4459$$

**Conclusión**: Alrededor del 44 % de los golpes de Tiger Woods viajan al menos entre 305 y 325 yardas.
