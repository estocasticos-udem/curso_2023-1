```python
from fractions import Fraction

def P(event, space): 
    "The probability of an event, given a sample space."
    return Fraction(cases(favorable(event, space)), 
                    cases(space))

favorable = set.intersection # Outcomes that are in the event and in the sample space
cases     = len              # The number of cases is the length, or size, of a set
```

## Ejemplo 1
Sea $X$ el lanzamiento de una moneda y $Y$ el lanzamiento de un dado:
* Determine $ùëÜ_ùëã$, $ùëÜ_ùëå$ y $ùëÜ_{ùëãùëå}$
* Dibuje la PMF conjunta ($ùëÉ_{ùëãùëå}(ùë•,ùë¶)$).
* Si se define un evento $ùê¥=\{ùëã+ùëå=3\}$ y $ùêµ=\{min‚Å°(ùëã,ùëå)=1\}$. Encuentre $ùëÉ[ùê¥]$ y $ùëÉ[ùêµ]$


Determine $ùëÜ_ùëã$, $ùëÜ_ùëå$ y $ùëÜ_{ùëãùëå}$

```python
Sx = [0, 1]   # Rango lanzamiento moneda
Sy = [1, 2, 3, 4, 5, 6]    # Rango lanzamiento de dados
Sxy = list(itertools.product(Sx, Sy))

# https://www.pythoncheatsheet.org/modules/itertools-module
Sx = [0, 1]
print(f"Sx: {Sx}") 
print(f"Sy: {Sy}") 
print(f"Sxy: {Sxy}") 
```

Dibuje la PMF conjunta ($ùëÉ_{ùëãùëå}(ùë•,ùë¶)$)

```python
# Empty dataframe
PMF = pd.DataFrame(columns = Sy, 
                   index = Sx)

# Llenado del dataframe
PMF = PMF.reset_index()
# print(PMF.index)
# print(PMF.columns)
print()
k = 0

for i in PMF.index:
    for j in PMF.columns[1:]:
        PMF.loc[i, j] = Fraction(1,len(Sxy))           
        k += 1
PMF
```

```python
# Grafica
import matplotlib.pyplot as plt
x = []
y = []
z = []
for i in range(len(Sx)):
    for j in range(len(Sy)):
        x.append(Sx[i])
        y.append(Sy[j])
        z.append(PMF.loc[Sx[i], Sy[j]])

print(x)
print(y)
print(z)
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
(markerline, stemlines, baseline) = ax.stem(x,y,z)
plt.setp(baseline, visible=False)
plt.show()
```

```python
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(6, 2))
sns.heatmap(table, annot=True, fmt=".3f", linewidths=.5, ax=ax)
ax.set_xlabel("Y")
ax.set_ylabel("X")
plt.show()
```

  |**$X=100$**|$0.20$|$0.10$|$0.20$|

  * La probabilidad de deducir $100 sobre ambas p√≥lizas.

## Ejemplo 4
Test two integrated circuits one after the other. On each test, the possible outcomes are $a$ ($accept$) and $r$ ($reject$). Assume that all circuits are acceptable with probability 0.9 and that the outcomes of successive tests are independent. Count the number of acceptable circuits $X$ and count the number of successful tests $Y$ before you observe the first reject. (If both tests are successful, let $Y = 2$). 
* Find the sample space of the experiment.
* Draw a tree diagram for the experiment.
* Find the VAs related with the experiment.
* Find the joint PMF of X and Y


Espacio muestran del experimento: $ùëÜ=\{ùëéùëé,ùëéùëü,ùëüùëé,ùëüùëü\}$:
* $P(aa) = P(a) \cdot P(a) = (0.9)(0.9) = 0.81$
* $P(ar) = P(a) \cdot P(r) = (0.9)(0.1) = 0.09$
* $P(ar) = P(r) \cdot P(a) = (0.1)(0.9) = 0.09$
* $P(ar) = P(r) \cdot P(r) = (0.9)(0.9) = 0.01$

Por otro lado las variables aleatorias $X$ y $Y$ se definen como:
* X = the number of acceptable circuits.
* Y = the number of successful tests before you observe the first reject.

De modo que:
* $S_X = {0,1,2}$
* $S_Y = {0,1,2}$

Luego tenemos que:

* $P(aa) = P(X=2,Y=2) = 0.81$
* $P(ar) = P(X=1,Y=1) = 0.09$
* $P(ra) = P(X=1,Y=0) = 0.09$
* $P(rr) = P(X=0,Y=0) = 0.01$

Vemos ademas que en terminos de estas VA, los casos que no aparecen es por que tienen probabilidad cero:
* P(X=1,Y=2) = 0$
* P(X=2,Y=1) = 0$

De modo que la PFM quedar√° como:

  |$P_{XY}(x,y)$|$Y=0$|$Y=1$|$Y=2$|
  |---|---|---|---|
  |**$X=0$**|$0.01$|$0$|$0$|
  |**$X=1$**|$0.09$|$0.09$|$0$|
  |**$X=2$**|$0$|$0$|$0.81$|

## Ejemplo 5

Lanzamiento de una moneda no cargada dos veces.

$$
\mathrm{First}=\begin{cases}
 & X=1\text{, } \mathrm{H}\\
 & X=0\text{, } \mathrm{T}
\end{cases} 
$$

$$
\mathrm{Second}=\begin{cases}
 & Y=1\text{, } \mathrm{H}\\
 & Y=0\text{, } \mathrm{T}
\end{cases} 
$$

Si $X$ y $Y$ son independientes, encontrar:
1. $P_{XY}(x,y)$
2. $F_{XY}(3,1)$

### Soluci√≥n punto 1

**Variables aleatorias**:
* **X**: Numero de caras del primer lanzamiento.
* **Y**: Numero de caras del segundo lanzamiento.

Luego tenemos que:

* $S_X = \{0,1\}$
* $S_Y = \{0,1\}$
* $S_{XY} = \{(0,0),(1,0),(1,0),(1,1)\}$

Todas la eventos son igualmente probables de modo que:
* $P_{XY}(X = 0, Y = 0)= 0.25$
* $P_{XY}(X = 1, Y = 0)= 0.25$
* $P_{XY}(X = 0, Y = 1)= 0.25$
* $P_{XY}(X = 1, Y = 1)= 0.25$



```python
# Rango de las variables aleatorias
Sx = [0, 1]
Sy = [0, 1]
Sxy = list(itertools.product(Sx,Sy))
print(f"Sx: {Sx}") 
print(f"Sy: {Sy}") 
print(f"Sxy: {Sxy}") 
```

```python
X = []
Y = []
for j in Sy:
    for i in Sx:
        X.append(i)
        Y.append(j)
Pxy = (len(Sx)*len(Sy))*[1/4]
df5 = pd.DataFrame({"X": X, "Y": Y, "Pxy": Pxy})
df5
```

```python
table5 = pd.pivot_table(df5, index=["X"], columns=["Y"], values=["Pxy"])
table5  # Pxy
```

```python
# PFM de X
column_names = list(table5.columns)
F_y = table5[column_names].sum(axis=0) # Suma de filas
F_y
```

```python
F_x = table5[column_names].sum(axis=1) # Suma de columnas
F_x
```

### Soluci√≥n punto 2

Tenemos que por definici√≥n:

$$ùêπ_{ùëãùëå}(ùë•,ùë¶)=ùëÉ(ùëã \leq ùë•,ùëå \leq ùë¶)=ùëÉ((ùëã \leq ùë•)ùëéùëõùëë(ùëå \leq ùë¶))$$

Como el resultado del primer lanzamiento es independiente del resultado del segundo lanzamiento tenemos:

$$ùêπ_{ùëãùëå}(ùë•,ùë¶)=ùëÉ(ùëã \leq ùë•)P(ùëå \leq ùë¶)=F_X(x) \cdot F_Y(y)$$

Aplicando esto en nuestro caso:

$$ùêπ_{ùëãùëå}(3,1)= F_X(3) \cdot F_Y(1)$$

Donde:
* $F_X(3) = \sum_{x\leq 3}P_X(X=x) =P_X(X=0) + P_X(X=1) + P_X(X=2) + P_X(X=3) = 0.5 + 0.5 + 0 + 0 = 1$
* $F_Y(1) = \sum_{x\leq 1}P_X(X=x) =P_X(X=0) + P_X(X=1) = 0.5 + 0.5 = 1$

Luego: 

$$ùêπ_{ùëãùëå}(3,1)= F_X(3) \cdot F_Y(1) = 1 \cdot 1 = 1$$

```python
F_x3 = sum(F_x.where(F_x <= 3))
F_y1 = sum(F_y.where(F_y <= 1))
F_x3_y1 = F_x3*F_y1
print(f"Fxy(3,1) = Fx(3)*Fy(1) = {F_x3:.2}*{F_y1:.2} = {F_x3_y1:.2} ")
```

## Ejemplo 6

Se dan dos variables aleatorias independientes $X \sim Bernoulli(p)$ y $Y \sim Bernoulli(q)$ $0 \le p,q \le 1$. Encontrar:
1. La **PFM** de $X$ y $Y$.
2. La **CDF** de $X$ y $Y$.

> **Variable aleatoria de $\bold{Bernoulli(p)}$**<br>
> Sea $X$ una variable aleatoria que sigue una distribuci√≥n Bernoulli con par√°metro $p$ $(X \sim Bernoulli(p))$, entonces la **PFM** ser√° de la forma:
> $$
> p(x) = P_X(x) = \begin{cases}
>  & 1 - p &, x = 0\\
>  & p &, x = 1\\
>  & 0 &, \text{eoc}\\
> \end{cases} 
> $$
>
> Por otro lado la PFM ser√°: 
> $$
> F(x) = F_X(x) = \begin{cases}
>  & 1 - p &, x = 0\\
>  & 1 &, x = 1\\
>  & 0 &, \text{eoc}\\
> \end{cases} 
> $$
> Donde $p$ es un parametro dentro del rango $0 < p < 1$

### Solucion punto 1 - PFM CONJUNTA

**Variables aleatorias**: 
* $X$ 
* $Y$

Luego como $X \sim Bernoulli(p)$ y $X \sim Bernoulli(q)$ entonces estas variables solo pueden tomar dos posibles valores:
* $S_X = \{0,1\}$
* $S_Y = \{0,1\}$
* $S_{XY} = \{(0,0),(1,0),(1,0),(1,1)\}$

Luego, teniendo en cuenta lo anterior y que $X$ y $Y$ son independientes:
* $P_{XY}(X = 0, Y = 0) = P_{X}(X = 0)P_{Y}(Y = 0) = (1-p)(1-q)$
* $P_{XY}(X = 0, Y = 1) = P_{X}(X = 0)P_{Y}(Y = 1) = (1-p)(q)$
* $P_{XY}(X = 1, Y = 0) = P_{X}(X = 1)P_{Y}(Y = 0) = (p)(1-q)$
* $P_{XY}(X = 1, Y = 1) = P_{X}(X = 1)P_{Y}(Y = 1) = pq$

De este modo la PDF ser√°:

$$
P_{XY}(x,y) = \begin{cases}
  & (1-p)(1-q) & , (x,y) =  (0,0)\\
  & (1-p)q & , (x,y) =  (0,1)\\
  & p(1-q) & , (x,y) =  (1,0)\\
  & pq & , (x,y) =  (1,1)\\
  & 0 & , \text{eoc}\\
\end{cases} 
$$

```python
# Rango de las variables aleatorias
Sx = [0, 1]
Sy = [0, 1]
Sxy = list(itertools.product(Sx,Sy))
print(f"Sx: {Sx}") 
print(f"Sy: {Sy}") 
print(f"Sxy: {Sxy}") 
```

```python
# Variables aleatorias

"""
Definici√≥n de la VA binomial: X ~ Bernoulli(p)

Parametros:
- p: 0.25
"""
p_ = 0.25
X = stats.bernoulli(p = p_)

x=[0,1]
pmf_x = X.pmf(x)
cdf_x = X.cdf(x)
df_x = pd.DataFrame({"x":x, "pmf":pmf_x, "cdf": cdf_x},columns=["x","pmf","cdf"])
df_x
```

```python
# Grafica de la PMF y la CDF de la VA X
fig, axs = plt.subplots(2)
fig.suptitle('PMF y CDF de $X \sim Bernoulli(p = 0.25)$')
# Grafica PFM
axs[0].plot(x, pmf_x, 'bs', ms=10)
axs[0].vlines(x, 0, pmf_x, colors='b', lw = 5, alpha=0.5)
axs[0].set_ylabel("$P_X(x)$")
# Grafica CDF
axs[1].plot(x, cdf_x, 'rs', ms=10)
axs[1].vlines(x, 0, cdf_x, colors='r', lw = 5, alpha=0.5)
axs[1].set_ylabel("$F_X(x)$")
axs[1].set_xlabel("x")
```

**Variables aleatorias**:

Luego como $X \sim Bernoulli(p)$ y $X \sim Bernoulli(q)$ entonces estas variables solo pueden tomar dos posibles valores:
* $S_X = \{0,1\}$
* $S_Y = \{0,1\}$
* $S_{XY} = \{(0,0),(1,0),(1,0),(1,1)\}$

Luego, teniendo en cuenta lo anterior y que $X$ y $Y$ son independientes:
* $F_{XY}(X = 0, Y = 0) = P_{XY}(X \leq 0, Y \leq 0) = P_{X}(X \leq 0)P_{Y}(Y \leq 0) = F_{X}(X \leq 0)F_{Y}(Y \leq 0) = (1-p)(1-q)$
* $F_{XY}(X = 0, Y = 1) = P_{XY}(X \leq 0, Y \leq 1) = P_{X}(X \leq 0)P_{Y}(Y \leq 1) = F_{X}(X \leq 0)F_{Y}(Y \leq 1) = (1)(1-p) = 1 - p$
* $F_{XY}(X = 1, Y = 0) = P_{XY}(X \leq 1, Y \leq 0) = P_{X}(X \leq 1)P_{Y}(Y \leq 0) = F_{X}(X \leq 1)F_{Y}(Y \leq 0) = (1)(1-q) = 1-q$
* $F_{XY}(X = 1, Y = 1) = P_{XY}(X \leq 1, Y \leq 1) = P_{X}(X \leq 1)P_{Y}(Y \leq 1) = F_{X}(X \leq 1)F_{Y}(Y \leq 1) = (1)(1) = $

De este modo la CDF ser√°:

$$
F_{XY}(x,y) = \begin{cases}
  & (1-p)(1-q) & , (x,y) =  (0,0)\\
  & 1-p & , (x,y) =  (0,1)\\
  & 1-q & , (x,y) =  (1,0)\\
  & 1 & , (x,y) =  (1,1)\\
  & 0 & , \text{eoc}\\
\end{cases} 
$$
